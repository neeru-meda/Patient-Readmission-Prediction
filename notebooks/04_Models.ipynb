```python
import joblib

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train = joblib.load("d:\\Codes\\Patient_Readmission_Prediction\\models\\x_train_balanced.pkl")
y_train = joblib.load("d:\\Codes\\Patient_Readmission_Prediction\\models\\y_train_balanced.pkl")

print("Training Shape :", x_train.shape)
print("Training Shape :", y_train.shape)
```

```python
lr = LogisticRegression(
    C = 1.0,
    class_weight = "balanced",
    random_state = 42,
    max_iter = 3000
)

lr.fit(x_train, y_train)

print("Logistic Regression Model Trained Successfully")
```

```python
dr = DecisionTreeClassifier(
    max_depth = 5,
    class_weight = "balanced",
    random_state = 42
)
dr.fit(x_train, y_train)
print("Decision Tree Model Trained Successfully")
```

```python
rf = RandomForestClassifier(
    n_estimators = 100,
    max_depth = 10,
    class_weight = "balanced",
    random_state = 42,
    n_jobs = -1
)
rf.fit(x_train, y_train)
print("Random Forest Model Trained Successfully")
```

```python
lr_model = joblib.load("d:\\Codes\\Patient_Readmission_Prediction\\models\\trained\\logistic_regression_model.pkl")
dr_model = joblib.load("d:\\Codes\\Patient_Readmission_Prediction\\models\\trained\\decision_tree_model.pkl")
rf_model = joblib.load("d:\\Codes\\Patient_Readmission_Prediction\\models\\trained\\random_forest_model.pkl")

print("type(lr_model) :", type(lr_model))
print("type(dr_model) :", type(dr_model))
print("type(rf_model) :", type(rf_model))
```
