```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
df = pd.read_csv('d:\\Codes\\Patient_Readmission_Prediction\\data\\raw\\diabetic_data.csv')
```

```python
df = df.replace('?', np.nan)
df_missing_values_columns = df.columns[df.isnull().any()]
print("Columns with missing values:")
print(df_missing_values_columns.tolist())
print("\nMissing values count for each column:")
total_missing = df[df_missing_values_columns].isnull().sum()
print(total_missing)
print("\nPercentage of missing values for each column:")
missing_percentage = (total_missing / len(df)) * 100
print(missing_percentage)
```

The columns that have the missing value percentage more than 40% are weight with 97% and payer_code with 40% , medical_specialty with 49% , max_glu_serum with 95% and A1Cresult with 83% 

```python
cols_to_drop = ['weight', 'payer_code', 'medical_specialty','encounter_id']
df = df.drop(columns=cols_to_drop)
print("Columns after dropping:")
print(df.columns.tolist())
```

```python
df['target'] = (df['readmitted']== '<30').astype(int)
df['target'].value_counts()
df = df.drop(columns=['readmitted'])
```

```python
# Keep only the most recent encounter per patient
df = df.drop_duplicates(subset='patient_nbr', keep='last')
```

```python
#Each age range is replaced by its midpoint.
age_map = {
    '[0-10)': 5,
    '[10-20)': 15,
    '[20-30)': 25,
    '[30-40)': 35,
    '[40-50)': 45,
    '[50-60)': 55,
    '[60-70)': 65,
    '[70-80)': 75,
    '[80-90)': 85,
    '[90-100)': 95
} 
df['age_numeric']= df['age'].map(age_map)
df = df.drop(columns=['age'])
```
