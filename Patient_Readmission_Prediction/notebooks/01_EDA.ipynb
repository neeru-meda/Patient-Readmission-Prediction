```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
x =df['readmitted'].value_counts().index
y = df['readmitted'].value_counts().values
plt.bar(x,y,color=['#2E6DA4','#E74C3C','#27AE60'])
plt.xlabel('Readmission Status')
plt.ylabel('Count')
plt.title('Distribution of Readmission Status')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
df['readmitted'].value_counts().plot(kind='bar', ax=ax, color=['#2E6DA4','#E74C3C','#27AE60'])
ax.set_title('Readmission Distribution (Class Imbalance Visible)')
ax.set_xlabel('Readmission Status')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('d:\\Codes\\Patient_Readmission_Prediction\\reports\\figures\\class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
readmit_by_age = df.groupby('age')['readmitted'].apply(lambda x: (x=='<30').sum()/len(x)*100)
readmit_by_age.plot(kind='bar', color="#A4A42E")
plt.title('Readmission Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Readmission Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('d:\\Codes\\Patient_Readmission_Prediction\\reports\\figures\\readmission_by_age.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns

numerical_cols = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'number_outpatient', 'number_emergency', 'number_inpatient','number_diagnoses']
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(18, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Numerical Features')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('d:\\Codes\\Patient_Readmission_Prediction\\reports\\figures\\correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
```
